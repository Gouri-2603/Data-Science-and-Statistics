```python
sizes = [30, 25, 20, 15, 10]
labels = ["A", "B", "C", "D", "E"]

# Custom colors (you can use named colors or hex codes)
colors = ["#FF9999", "#66B2FF", "#99FF99", "#FFCC99", "#C2C2F0"]

# Create pie chart
plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct="%1.1f%%",
    startangle=90
)

# Add title
plt.title("Pie Chart with Custom Colors")

plt.show()
```

```python
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=20, color="#69b3a2", edgecolor="black")

# Add title and labels
plt.title("Basic Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
